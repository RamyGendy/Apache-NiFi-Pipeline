```python
null_values=df[df.isnull().any(axis=1)]
null_phone_no=df[df['Phone_Number'].isnull()]
null_product_margin=df[df['Product_Base_Margin'].isnull()]
null_Email=df[df['Email'].isnull()]
```

```python
# investigating Product_Category values
df['Product_Category'].value_counts()
```

```python
# investigating Product_Sub_Category values
df['Product_Sub_Category'].value_counts()
```

```python
# explore Product_Category values
df['Product_Category'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%', label='');
plt.title('Product Category');
```

```python
new_column_list= ['Ord_id', 'Prod_id', 'Ship_id', 'Sales',
       'Discount', 'Order_Quantity', 'Profit', 'Shipping_Cost',
       'Product_Base_Margin', 'Order_ID', 'Order_Date', 'Order_Priority',
                  'Product_Sub_Category', 'Ship_Mode', 'Ship_Date']
```

```python
# sales department - office supplies suggested view
off_supplies = df[df.Product_Category == 'OFFICE SUPPLIES'][new_column_list]
off_supplies.head()
```

```python
# sales department - furniture suggested view
furn = df[df.Product_Category == 'FURNITURE'][new_column_list]
furn.head()
```

```python
# sales department - technology suggested view
tech = df[df.Product_Category == 'TECHNOLOGY'][new_column_list]
tech.head()
```
